```python
import pandas as pd
```

```python
df = pd.read_csv('corruption.csv')
```

```python
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.scatterplot(data = df, x = 'socialnetshare', y = 'corruption')
plt.title('Corruption Vs Social Media Penetration')
plt.show()
```

### It's showing inverse relationship between corruption and socialnetshare, i.e; Corruption decreases as socialnetshare increases

```python
# Calculate quintiles
df['GDP_quintile'] = pd.qcut(df['lngdp'], 5, labels=['Q1', 'Q2', 'Q3', 'Q4', 'Q5'])

# Separate top and bottom quintiles
df_top_bottom = df[df['GDP_quintile'].isin(['Q1', 'Q5'])]
```

```python
# Scatter plot for top and bottom GDP quintiles
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_top_bottom, x='socialnetshare', y='corruption', hue='GDP_quintile')
plt.title('Top and Bottom Quintiles: Social Media Penetration vs Corruption')
plt.xlabel('Social Media Penetration')
plt.ylabel('Corruption Index')
plt.legend(title='GDP Quintile')
plt.show()
```

```python
### From the visuals we can say that the top quintile(Q1) data has more corruption with less social media penetration, but there is one data point which is not following this relation 
### While the Boottom quintile(Q5) has less corruption with more Social Media Pene
```
